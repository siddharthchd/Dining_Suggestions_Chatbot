```python
import pandas as pd

df = pd.read_csv('restaurants_yelp.csv', header = None)
df.columns = ['restaurantId', 'restaurantName', 'cuisineType', 'restaurantAddress', 
          'Longitude', 'Latitude', 'restaurantRating', 'reviewCount', 'priceRange', 'insertedAtTimestamp']
print(len(df))
print(df.dtypes)
df.head()
```

```python
import re

df['restaurantId'] = df['restaurantId'].str.replace('-', '')
df['restaurantId'] = df['restaurantId'].str.replace(' ', '')
df['restaurantId'] = df['restaurantId'].str.replace('_', '')
df['restaurantId'].head(50)
```

```python
#df['restaurantId'].replace('-', '', inplace = True)
df = df.drop_duplicates(subset = 'restaurantId')
#df.drop_duplicates('restaurantId', inplace = True)
#df.set_index('restaurantId', inplace = True)
df['priceRange'].fillna('$$', inplace = True)
df.head(50)
```
